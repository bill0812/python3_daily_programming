# 簡易技術分析

當短週期均線(20MA)由下方往上突破長週期均線(60MA)時，即為『黃金交叉』。
反之，若短週期均線(20MA)由上方往下跌破長週期均線(60MA)時，為『死亡交叉』。
一般而言，黃金交叉表示短期內走多，視為買進點；而死亡交叉表示短期內走空，視為賣出點。
聲明：均線只是一種指標，不代表照著操作就能獲利，還是要考量多種因素再下決定。

```python
df=web.DataReader("AAPL","yahoo",start="2017-1-1",end="2018-11-30")
```

```python
p=df["Adj Close"]
```

```python
MA20 = p.rolling(window=20).mean()
MA60 = p.rolling(window=60).mean()
```

```python
# 判斷黃金交叉或是死亡交叉
```

```python
for i in range(len(MA60)-60):
    if MA20[i+59]<MA60[i+59] and MA20[i+60]>MA60[i+60] : #MA60在[59]才有值
        print(df.index[i+60],'黃金交叉')
    if MA20[i+59]>MA60[i+59] and MA20[i+60]<MA60[i+60] :
        print(df.index[i+60],'死亡交叉')
```
