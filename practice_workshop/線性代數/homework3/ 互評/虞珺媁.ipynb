```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sympy as s
import numpy.random as rn
```

```python
from scipy.interpolate import *
```

```python
x=np.linspace(1,20,15)
y=2*x+5*np.sin(x)+7*np.log(x**2)
plt.plot(x,y,'o',label='data')
plt.xlabel('x')
plt.ylabel('y')
```

```python
x=np.linspace(1,20,15)
y=2*x+5*np.sin(x)+7*np.log(x**2)

x1=np.linspace(1,20,15)
y1=2*x+5*np.sin(x)+7*np.log(x**2)

plt.plot(x,y,'o',label='data')
plt.plot(x1,y1,':',label='prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
```
