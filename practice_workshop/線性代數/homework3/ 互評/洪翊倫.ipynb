```python
import numpy as np
from scipy.interpolate import lagrange
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,195,87,78]
y = [2,np.pi,50,40,-2,20]
```

```python
g =lagrange(x, y)
```

```python
X = np.linspace(0,200,10000)
Y = g(X)
```

```python
plt.plot(X,Y)
plt.scatter(x,y,c ="b")
plt.ylim()
```
