```
import matplotlib.pyplot as plt
import numpy as np

def Lagrange(points, x):
  sum = 0
  n = len(points)
  for i in range(n):
    xi, yi = points[i]
    def L(i):
      Lvalue = 1
      for j in range(n):
        if i==j:
          continue
        xj, yj = points[j]
        Lvalue *= float(x-xj)/float(xi-xj)
      return Lvalue
    sum +=yi*L(i)
  return sum

mydata = np.array([[0,0],[1,1],[3,9],[5,25],[6,40],[8,64],[10,100],[12,144]])
z=[]
for i in range(13):
    z.append([1,np.round(Lagrange(mydata,1),10)])

z=np.array(z)

print(z)

plt.plot(mydata[:,0],mydata[:,1], 'r', z[:,0], z[:,1], 'b')

plt.show()      
    

```
