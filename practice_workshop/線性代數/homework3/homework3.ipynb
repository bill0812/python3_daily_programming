```python
# 固定（設定） x , y 座標得值
x_fixed = np.array([1,2,3])
y_fixed = np.array([2,1,3])

# 用上面固定的座標點找出拉格朗日插值法的係數方程式
g_fixed = lagrange(x_fixed,y_fixed)
```

```python
# 在 0 與 5 間分成 10000 個點，並設成 X 軸
X_fixed = np.linspace(0,5,10000)

# 將 X_fixed 的值代入此拉格朗日插值法方程式中，並找出 Y 點
Y_fixed = g_fixed(X_fixed)

# 畫出 g_fixed，並將 x_fixed,y_fixed 畫在圖中
plt.plot(X_fixed,Y_fixed)
plt.scatter(x_fixed,y_fixed, c="b")
plt.ylim(0,8)

# 印出係數
print("g_fixed 係數 : " +str(g_fixed.c))
```
