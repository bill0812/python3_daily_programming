```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sympy as s
```

```python
X = [0, 0]
Y = [0, 0]
U = [2, 3]
V = [1, 1]
```

```python
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

```python
plt.quiver(X, Y, U, V, color=['g', 'b'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

```python
X = [0]
Y = [0]
U = [2, 3, 3]
V = [1, 1, 4]
plt.quiver(X, Y, U, V, color=['g', 'b', 'r'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-5, 10)
plt.ylim(-5, 10)
```
