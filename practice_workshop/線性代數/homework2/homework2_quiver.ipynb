```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sympy as s
```

```python
# easy try for simple quiver begin at (0,0)

# orgin x and y use arg* pattern to use x and y of origin point
origin0 = [0], [0]

# other simple way to use x and y of origin point
x0 = [-1, 0]
y0 = [-1, 0]

# u0 represent all x axis of vector ; v0 represent all axis y of vecto
u0 = [3, -3]
v0 = [2, -1]

plt.quiver(x0, y0, u0, v0, color=['g', 'b'], angles='xy', scale_units='xy', scale=1)
#plt.quiver(*origin0, u0, v0, color=['g', 'b'], angles='xy', scale_units='xy', scale=1)

plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

```python
# use another way to show some vector on x-y plane
x_meshgird = np.arange(-1, 1, 1)
y_meshgird = np.arange(-1, 1, 1)
print("X start : " + str(x_meshgird) + "\nY start : " + str(y_meshgird))

u_meshgird, v_meshgird = np.meshgrid(x_meshgird, y_meshgird)
print("\nX end : \n" + str(u_meshgird) + "\n\nY end : \n" + str(v_meshgird))

# notice : from https://matplotlib.org/examples/pylab_examples/quiver_demo.html
# If X and Y are absent, they will be generated as a uniform grid. 
# If U and V are 2-D arrays and X and Y are 1-D, 
# and if len(X) and len(Y) match the column and row dimensions of U, 
# then X and Y will be expanded with numpy.meshgrid().

plt.quiver(x_meshgird, y_meshgird, u_meshgird, v_meshgird, color=['g', 'b' , 'y' , "orange"], angles='xy', scale_units='xy', scale=1)
plt.xlim(-3, 1)
plt.ylim(-3, 1)
```

```python
# use another function : quiverkey
x_quiverkey = np.arange(-1, 1, 1)
y_quiverkey = np.arange(-1, 1, 1)
u_quiverkey, v_quiverkey = np.meshgrid(x_quiverkey, y_quiverkey)

# combin fig and ax in one line using subplot
# fig_first , ax_first = plt.subplots()

fig_first = plt.figure()

# add first axe to figure
ax_first = fig_first.add_subplot(2,1,1)

# add second axe to figure
# ax_seond = fig_first.add_subplot(2,1,1)
q = ax_first.quiver(x_quiverkey, y_quiverkey, u_quiverkey, v_quiverkey, color=['g', 'b' , 'y' , "orange"], angles='xy', scale_units='xy', scale=1)

# using quiverkey to add a key on the plot
ax_first.quiverkey(q, X=0.15, Y=0.15, U=15,
             label='Quiver key, length = 10', labelpos='E')

# set first axe x , y limit
ax_first.set_xlim(-3, 1)
ax_first.set_ylim(-3, 1)
```

```python
# from upper try , use more dimension matrix to plot
x_quiverkey_2 = np.arange(-10, 10, 1)
y_quiverkey_2 = np.arange(-10, 10, 1)
u_quiverkey_2, v_quiverkey_2 = np.meshgrid(x_quiverkey_2, y_quiverkey_2)

# same way to plot a more complicated quiver plot
fig_second, ax_second = plt.subplots()
q = ax_second.quiver(x_quiverkey_2, y_quiverkey_2, u_quiverkey_2, v_quiverkey_2, color=['g', 'b' , 'y' , "orange"], angles='xy', scale_units='xy', scale=1)
ax_second.quiverkey(q, X=0.5, Y=0.5, U=10,
             label='Quiver key, length = 10', labelpos='E')

plt.show()
```

```python
# try to give a random int number c to multiply [x,y]
# first should use transfer matrix to change format
# we use first tset x0 = [-1, 0] , y0 = [-1, 0] , u0 = [3, -3] , v0 = [2, -1]

# use column_stack function to change all seperated [x1,x2,...] and [y1,y2,...,]  to [x,y]
S_new_1 = np.column_stack((u0, v0))
print("S_new 1 : \n" + str(S_new_1))
print("S_new 1's shape : " + str(S_new_1.shape[0]))

# random init a variable int c 
c_1 = np.random.randn(1,S_new_1.shape[0])
print("c 1: \n" + str(c_1))

# mutiply change and the original vector
mutiply_result = np.dot(c_1,S_new_1)
mutiply_result = mutiply_result.astype(int)
print("\nmutiply result : \n" + str(mutiply_result))

# add result to u0 , v0
u0.append(list(mutiply_result)[0][0])
v0.append(list(mutiply_result)[0][1])
print("\nCurrent u0 : " + str(u0))
print("\nCurrent v0 : " + str(v0))

# put new u0 and v0 to quiver and plot
# plt.quiver(x0, y0, u0, v0, color=['g', 'b'], angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin0, u0, v0, color=['g', 'b' , "y" , "orange" , "purple" ], angles='xy', scale_units='xy', scale=1)

plt.xlim(min(u0), max(u0))
plt.ylim(min(v0), max(v0))
```

```python
# from upper try , use more dimension matrix to plot
np.seterr(divide='ignore', invalid='ignore')
origin1 = [0], [0]
x_quiverkey_3 = np.arange(-5, 5, 1)
y_quiverkey_3 = np.arange(-5, 5, 1)
u_quiverkey_3, v_quiverkey_3 = np.meshgrid(x_quiverkey_3, y_quiverkey_3)

# modify u_quiverkey_3 , v_quiverkey_3 from 2d to 1d
u_quiverkey_3_changed = list(u_quiverkey_3.flatten())
v_quiverkey_3_changed = list(v_quiverkey_3.flatten())

S_new_2 = np.column_stack((u_quiverkey_3_changed, v_quiverkey_3_changed))
print("S_new 2 : \n" + str(S_new_2))
print("S_new 2's shape : " + str(S_new_2.shape[0]))

# random init a variable int c 
c_2 = np.random.randn(1,S_new_2.shape[0])
print("c 2: \n" + str(c_2))

# mutiply change and the original vector
mutiply_result = np.dot(c_2,S_new_2)
mutiply_result = mutiply_result.astype(int)
print("\nmutiply result : \n" + str(mutiply_result))

# add result to u0 , v0
u_quiverkey_3_changed.append(list(mutiply_result)[0][0])
v_quiverkey_3_changed.append(list(mutiply_result)[0][1])
print("\nCurrent u0 : " + str(u_quiverkey_3_changed))
print("\nCurrent v0 : " + str(v_quiverkey_3_changed))

# same way to plot a more complicated quiver plot
fig_second, ax_second = plt.subplots()
q = plt.quiver(*origin1, u_quiverkey_3_changed, v_quiverkey_3_changed, color=['g', 'b' , "y" , "orange" , "purple" ], angles='xy', scale_units='xy', scale=1)
ax_second.quiverkey(q, X=0.5, Y=0.5, U=10,
             label='Quiver key, length = 10', labelpos='E')

plt.xlim(min(u_quiverkey_3_changed), max(u_quiverkey_3_changed))
plt.ylim(min(v_quiverkey_3_changed), max(v_quiverkey_3_changed))
plt.show()
```
